## Effects of different kernels values as feature extractor

```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
# Fetching the image using its URL
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
# The image experimented on
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
```

```
# The below image shows all the edges combined

cv2_imshow(edges)
```

## Note:
### The sum of all elements of kernel matrix is 1. If the sum is not exactly one, the resultant image will be brighter or darker.

### Vertical edge detector

```
# The following kernel is to detect vertical egdes

kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

# Applying kernel on image
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### Horizontal edge detector

```
# The following kernel is to detect horizontal egdes

kernel = np.float32([[1,1,1],[0,0,0],[-1,-1,-1]])

# Applying kernel on image
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### 45 deg edge **detector**

```
# The following kernel is to detect 45 deg angled egdes

kernel = np.float32([[-1,-1,0],[-1,0,1],[0,1,1]])

# Applying kernel on image
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### 135 deg edge detector

```
# The following kernel is to detect 135 deg angled egdes

kernel = np.float32([[0,1,1],[-1,0,1],[-1,-1,0]])

# Applying kernel on image
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### Blur Kernel

```
# The following kernel is to blur/smoothening the image
# Also called as averaging due to the kernel values chosen
# This kernel un-focus on the difference in adjacent pixels.

kernel = np.float32([[1/15,1/15,1/15],[1/15,1/15,1/15],[1/15,1/15,1/15]]) 

# Applying kernel on image
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### Sharpen Kernel

```
# The following kernel is to sharpen the image

# This kernel focus on the difference in adjacent pixels.
kernel = np.float32([[0,-5,0],[-5,21,-5],[0,-5,0]])

# Applying kernel on image
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

### Identity Kernel

```
# The following kernel is an identity which return the image as is.

# Returns the same image without doing anything
kernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])

# Applying kernel on image
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
